```python
import pandas as pd
```

```python
df = pd.read_csv("OSX_DS_assignment.csv")
```

```python
# To derive the top 5 actionable Insights from the Data
```

```python
# 1) Popular Wine Varieties 
pop = df['variety'].value_counts().head(5)
pop
```

```python
# 2) Scatter plot of points vs. price
#    Helps in analyzing the insights on the price-value proposition
import matplotlib.pyplot as plt
plt.scatter(df['points'], df['price'])
plt.xlabel('Points')
plt.ylabel('Price')
plt.title('Points vs. Price')
plt.show()
```

```python
# 3) Count the number of wines by country
wine_counts = df['country'].value_counts()
wine_counts.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Number of Wines')
plt.title('Distribution of Wines by Country')
plt.show()
```
